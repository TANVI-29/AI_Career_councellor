```python
df["interest_text"]=df["interest_text"].str.lower()
```

```python
import string
exclude= string.punctuation
def rem_punc(text):
    return text.translate(str.maketrans('','',exclude))

df["interest_text"]=df["interest_text"].apply(rem_punc)
```
